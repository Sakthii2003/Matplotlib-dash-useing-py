```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#step1: load csv using pandas
df = pd.read_csv(r"D:\DATA analyst\sales_data.csv")

#step2: convert columns to numpy arrays
days=df['Day'].to_numpy()
sales_a = df["Product_A_Sales"].to_numpy()
sales_b=df['Product_B_Sales'].to_numpy()

#step3:display data
print("First 5 row:\n",df.head())
```

```python
#step 1:plot the sales data 
plt.figure(figsize=(8,5))
plt.plot(days,sales_a,label="product A",marker=r"$\heartsuit$",linestyle='-.',color='#0b4f09')
plt.plot(days,sales_b,label="product b",marker=r"$\heartsuit$",linestyle='-.',color='#650a73')
plt.title('sales Comparision:Product A vs Product B ')
plt.xlabel('day')
plt.ylabel('sales')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
#2. Stacked Area Chart
plt.figure(figsize=(8, 5))
plt.stackplot(days, sales_a, sales_b, labels=['Product A', 'Product B'], colors=['#0b4f09', '#650a73'])
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Stacked Area Chart: Product A and B Sales')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#4. Subplot: Line Charts Side by Side
# Product A
ig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(days, sales_a, marker='o', color='#0b4f09')
axs[0].set_title('Product A Sales')
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Sales')
axs[0].grid(True)

# Product B
axs[1].plot(days, sales_b, marker='s', color='#650a73')
axs[1].set_title('Product B Sales')
axs[1].set_xlabel('Day')
axs[1].set_ylabel('Sales')
axs[1].grid(True)

plt.tight_layout()
plt.show()

#You can try all of these by replacing your plt.plot(...) section with any of the above.
```

```python
#3. Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(days, sales_a, color='#0b4f09', label='Product A', marker='o')
plt.scatter(days, sales_b, color='#650a73', label='Product B', marker='s')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Scatter Plot: Sales of Products A & B')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
bar_width=0.35
x = np.arange(len(days))

plt.bar(x, sales_a, width=bar_width, label='Product A', color='#0b4f09')
plt.bar(x + bar_width, sales_b, width=bar_width, label='Product B', color='#650a73')

plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Bar Chart: Product A vs B Sales')
plt.xticks(x + bar_width / 2, days)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Step 4: Plot the sales data

Day = df['Day']
sales_a = df['Product_A_Sales']
sales_b = df['Product_B_Sales']
plt.figure(figsize=(8, 5))
plt.plot(Day, sales_a, label='Product A', marker='o', color='#0b4f09')
plt.plot(Day, sales_b, label='Product B', marker='s', color='#650a73')
plt.title('Sales Comparison: Product A vs Product B')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
